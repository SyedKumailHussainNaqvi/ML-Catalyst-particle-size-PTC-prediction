```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score,KFold
from xgboost import XGBRegressor
```

```python
data_input=pd.read_csv(r"C:\Users\user\Energy_AI\ML-Metal-HER-Catalyst-main\ML-Metal-HER-Catalyst-main\HER-catalyst-data-with-DOI.csv", sep=',')
```

```python

labels=data_input['Overpotential (mV)']
features=data_input.drop('Overpotential (mV)', axis=1).drop('DOI', axis=1).drop('Composition', axis=1)

X_train,X_test,y_train,y_test=train_test_split(features, labels, test_size=0.2, random_state=42) #42
```

```python
train_data = pd.concat([X_train, pd.DataFrame(y_train, columns=['Overpotential (mV)'])], axis=1)
train_data
```

```python
file_path = "train_data.csv"
train_data.to_csv(file_path, index=False)
print(f"Test data saved to {file_path}")
```

```python
test_data = pd.concat([X_test, pd.DataFrame(y_test, columns=['Overpotential (mV)'])], axis=1)
test_data
```

```python
file_path = "test_data.csv"
test_data.to_csv(file_path, index=False)
print(f"Test data saved to {file_path}")
```
